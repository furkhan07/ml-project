```python
num_cols = X.select_dtypes(include='number').columns
obj_cols = X.select_dtypes(exclude='number').columns
```

#it reduce no of preprocessing steps        
#helps to reduce number of steps

```python
compose = ColumnTransformer(
    transformers=[
        ('scaling',StandardScaler(),num_cols),  #we cant add multiple preprocessing for numerical in this method - BY USING PIPELINES WE CAN DO THAT
        ('encoding',OneHotEncoder(),obj_cols)
    ]
)
```

#important parameter - transformer and remainder       
`Remainder` -     
drop - drop the column          
pass through - keep the column   ( skip and keep the columns whcih are mentioned in num_cols or obj_cols) ( any column is missed it keeps)

# Parameters

```python
compose = ColumnTransformer(
    transformers=[
        ('scaling',StandardScaler(),num_cols),  #we cant add multiple preprocessing for numerical in this method - BY USING PIPELINES WE CAN DO THAT
        ('encoding',OneHotEncoder(),obj_cols)
    ],remainder='drop'
)
compose.fit_transform(df) #1.      
```

```python
compose = ColumnTransformer(
    transformers=[
        ('scaling',StandardScaler(),num_cols),  #we cant add multiple preprocessing for numerical in this method - BY USING PIPELINES WE CAN DO THAT
        ('encoding',OneHotEncoder(),obj_cols)
    ],remainder='passthrough'
)
compose.fit_transform(df)  # 1.0, 'Yes'
```

```python
xtrain = compose.fit_transform(xtrain)
xtest = compose.transform(xtest)
```
